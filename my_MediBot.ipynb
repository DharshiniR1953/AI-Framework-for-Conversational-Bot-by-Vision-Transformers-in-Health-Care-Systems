```python
from langchain.chains import RetrievalQA
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.llms import LlamaCpp
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.document_loaders import PyPDFDirectoryLoader
```

```python
#load pdf files
loader = PyPDFDirectoryLoader("/drive/My Drive/data/")
data = loader.load()
```

```python
data
```

```python
#Step 05: Split the Extracted Data into Text Chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=20)

text_chunks = text_splitter.split_documents(data)
```

```python
len(text_chunks)
```

```python
#get the third chunk
text_chunks[2]
```

```python
#Step 06:Downlaod the Embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
#Step 08: Create Embeddings for each of the Text Chunk
vector_store = FAISS.from_documents(text_chunks, embedding=embeddings)
```

```python
#Import Model
llm = LlamaCpp(
    streaming = True,
    model_path="/drive/My Drive/Trained_Models/mistral-7b-instruct-v0.1.Q4_K_M.gguf",
    temperature=0.75,
    top_p=1,
    verbose=True,
    n_ctx=4096
)
```

method one

```python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vector_store.as_retriever(search_kwargs={"k": 3}))
```

```python
import sys

while True:
  user_input = input(f"Input Prompt: ")
  if user_input == 'exit':
    print('Exiting')
    sys.exit()
  if user_input == '':
    continue
  result = qa({'query': user_input})
  print(f"Answer: {result['result']}")
```

method two


```python
from langchain.chains.question_answering import load_qa_chain
```

```python
chain=load_qa_chain(llm=llm, chain_type="stuff")
```

```python
query="What is turmeric?"
```

```python
docs = vector_store.similarity_search(query=query, k=3)
```

```python
chain.run(input_documents=docs, question=query)
```
